***MACHINE LEARNING*** MEAN(), MEDIAN(), MODE().***

```python
import numpy

speed = [99,88,55,44,66,33,22,11,77,54,55,62,53,64,95]

x = numpy.mean(speed)

print(x)
```

```python

import numpy

speed = [99,88,55,44,66,33,22,11,77,54,55,62,53,64,95]

x = numpy.median(speed)

print(x)
```

```python
from scipy import stats

speed = [99,88,55,44,66,33,22,11,77,54,55,62,53,64,95]

x = stats.mode(speed)

print(x)
```
